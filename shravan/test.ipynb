```python
import matplotlib.pyplot as plt
```

```python
df['trending'].value_counts().plot(kind='bar')
plt.show()
```

```python
trend['trending_date'] = pd.to_datetime(trend['trending_date'], format='%Y-%m-%d')
trend['publishedAt'] = pd.to_datetime(trend['publishedAt'], format='%Y-%m-%d')

views_by_date = trend.groupby('trending_date')['views per day'].mean()
plt.plot(views_by_date.index, views_by_date.values)
plt.title('Trend of Views per Day for Trending Videos')
plt.xlabel('Trending Date')
plt.ylabel('Views per Day')
plt.show()
```

```python
grouped = trend.groupby('trending_date').mean()
dates = grouped.index

plt.plot(dates, grouped['views per day'], label='Views per Day')
plt.plot(dates, grouped['likes per day'], label='Likes per Day')
plt.plot(dates, grouped['dislikes per day'], label='Dislikes per Day')
plt.plot(dates, grouped['comments per day'], label='Comments per Day')
plt.title('Trend of Engagement Metrics per Day for Trending and Non-trending Videos')
plt.xlabel('Date')
plt.ylabel('Metric Value')
plt.legend()
plt.show()
```
