```python
df = pd.concat([nontrend, trend], ignore_index=True)
```

```python
data = df.to_numpy()
np.random.shuffle(data)
X = data[:, (5, 8, 9, 10, 11, 13)].astype(np.float64)
y = data[:, 15].astype(np.float64)
X = X - np.mean(X, axis=0, keepdims=True)
X = X / np.std(X, axis=0, keepdims=True)
print(X)
print(y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.datasets import make_moons, make_circles, make_classification
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.inspection import DecisionBoundaryDisplay

names = [
    "Nearest Neighbors",
    # "Linear SVM",
    # "RBF SVM",
    # "Gaussian Process",
    "Decision Tree",
    "Random Forest",
    # "Neural Net",
    "AdaBoost",
    "Naive Bayes",
    "QDA",
]

classifiers = [
    KNeighborsClassifier(3),
    # SVC(kernel="linear", C=0.025),
    # SVC(gamma=2, C=1),
    # GaussianProcessClassifier(1.0 * RBF(1.0)),
    DecisionTreeClassifier(max_depth=5),
    RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1),
    # MLPClassifier(alpha=1, max_iter=1000),
    AdaBoostClassifier(),
    GaussianNB(),
    QuadraticDiscriminantAnalysis(),
]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

for name, classifier in zip(names, classifiers):
    start = time.time()
    classifier.fit(X_train, y_train)
    end = time.time()
    t = end - start
    train_score = classifier.score(X_train, y_train)
    test_score = classifier.score(X_test, y_test)
    print(name, "took", t, "s to run, train accuracy:", train_score, "test_accuracy:", test_score)
```
