```python
# tesing sklearn
import numpy as np
from sklearn.pipeline import make_pipeline
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
X = np.array([[-1, -1], [-2, -1], [1, 1], [2, 1]])
print(np.shape(X))
y = np.array([1, 1, 2, 2])
print(np.shape(y))
from sklearn.svm import SVC
clf = make_pipeline(StandardScaler(), SVC(gamma='auto'))
clf.fit(X, y)
Pipeline(steps=[('standardscaler', StandardScaler()),
                ('svc', SVC(gamma='auto'))])


print(clf.predict([[-0.8, -1]]))
```

```python

# The View Column
X = df['views per day'].to_numpy().reshape(-1,1)
# Trending Label
y = df['trending'].to_numpy()
print(np.shape(X))
print(np.shape(y))

clf = make_pipeline(StandardScaler(), SVC(gamma='auto'))
clf.fit(X, y)
Pipeline(steps=[('standardscaler', StandardScaler()),
                ('svc', SVC(gamma='auto'))])

testVideo = [df.to_numpy()[-1, 8]]
actual = [df.to_numpy()[-1, -1]]
predicted =  clf.predict([testVideo])
print("actual: ", actual)
print("predicted; ", predicted)
```

```python
#non trending
points = df.to_numpy()[np.where(y == 0)[0]]
plt.plot(points[:, 8], points[:, -1],'x', color="blue", label = 'trending')

#trending
points = df.to_numpy()[np.where(y == 1)[0]]
plt.plot(points[:, 8], points[:, -1],'x', color="red", label = 'non trending')

#testing
plt.plot([testVideo, testVideo], [actual, predicted],'x', color="black", label = 'testing')

plt.legend()
```
